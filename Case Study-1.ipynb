```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cars_data.csv')
```

```python
data
```

```python
#checking for null values in the dataset
data.isnull().values.any()
```

```python
#Plot a bar graph of male vs female buyers participated in the sales.
```

```python
#count the no of male and female buyers in the datset
data['Buyer Gender'].value_counts()
```

```python
#plt.figure(figsize=(4,4))
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(4,4), rot=0)
plt.xlabel("Gender", labelpad=9)
plt.ylabel("Count of People", labelpad=9)
plt.title("Count of  Male and Female Buyers ", y=1.02)
```

```python
 # the top 5 cars based on their sales price.
```

```python
sorted_data = data.sort_values(by=["Sale Price"], ascending=False)
sorted_data.head(5)
```

```python
# Inference: Top 5 cars based on sale price are Volkswagen,Audi,Lexus,Mazda,Lexus
```

```python
#the least 5 cars based on their Resell price.
```

```python
sorted_data = data.sort_values(by=["Resell Price"], ascending=True)
sorted_data.head(5)
```
